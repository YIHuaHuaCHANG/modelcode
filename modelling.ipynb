```python
import pulp
from pulp import GLPK
import pandas as pd # Library to handle data
import numpy as np
data=pd.read_excel("C:/Users/USER/Desktop/Modelling/modelling.xlsx")
```

```python
def fill(x):
    results = x.fillna(x.mean(), inplace=True)
    return results
fill(data["Percentage modal split for commuting by staff (bus)"])
fill(data["total water consumption (m3)"])
fill(data["total volume of waste water(m3)"])
fill(data["Percentage modal split for commuting by students (bus)"])
data
```

```python

data1 = pd.DataFrame()

for index, row in data.iterrows():
    
    max_area = data["total site area (hectares)"].max()
    max_staff = data["Percentage modal split for commuting by staff (bus)"].max()
    max_students = data["Percentage modal split for commuting by students (bus)"].max()
    max_teaching = data["Teaching students headcount"].max()
    max_research = data["Research students headcount"].max()
    max_lecture = data["public lectures"].max()
    max_funds = data["UK government regeneration funds"].max()
    max_energy_comsumption = data["Total energy consumption(kWh)"].max()
    max_water_consumption = data["total water consumption (m3)"].max()
    max_waste_water = data["total volume of waste water(m3)"].max()
    out_setting= data["setting and infrastructure"].max()
    out_energy= data["energy and climate change"].max() 
    out_waste= data["waste"].max()
    out_water= data["water"].max()
    out_transportation= data["transportation"].max()
    out_education= data["education and research"].max()
    
    #positive
    row["total site area (hectares)"] = row["total site area (hectares)"] / max_area
    row["Percentage modal split for commuting by staff (bus)"] = row["Percentage modal split for commuting by staff (bus)"] / max_staff
    row["Percentage modal split for commuting by students (bus)"] = row["Percentage modal split for commuting by students (bus)"] / max_students
    row["Teaching students headcount"] = row["Teaching students headcount"] / max_teaching
    row["Research students headcount"] = row["Research students headcount"] / max_research
    row["public lectures"] = row["public lectures"] / max_lecture
    row["UK government regeneration funds"] = row["UK government regeneration funds"] / max_funds
    
    row["setting and infrastructure"] = row["setting and infrastructure"]/out_setting
    row["energy and climate change"] = row["energy and climate change"]/out_energy
    row["waste"] = row["waste"]/out_waste
    row["water"] = row["water"]/out_water
    row["transportation"] = row["transportation"]/out_transportation
    row["education and research"] = row["education and research"]/out_education
    
    #negative
    row["Total energy consumption(kWh)"] = 1 - (row["Total energy consumption(kWh)"] / max_energy_comsumption)
    row["total water consumption (m3)"] = 1 - (row["total water consumption (m3)"] / max_water_consumption)
    row["total volume of waste water(m3)"] = 1 - (row["total volume of waste water(m3)"] / max_waste_water)
    
    data1 = data1.append(row)

print(data1)
```

```python
for k in range(7):  # given that 4 hospitals (DMUs) exist, we expect to solve 4 LP problems to get the most favourable relative efficiency score for each hospital
    model = pulp.LpProblem("University environmental Analytics", pulp.LpMaximize) # Create an LP maximization problem
    m1 = pulp.LpVariable("m1", lowBound=0, upBound=None, cat='Continuous') # Create a variable m1>=0
    n1 = pulp.LpVariable("n1", lowBound=0, upBound=None, cat='Continuous') # Create a variable n1>=0
    n2 = pulp.LpVariable("n2", lowBound=0, upBound=None, cat='Continuous') # Create a variable n2>=0
    n3 = pulp.LpVariable("n3", lowBound=0, upBound=None, cat='Continuous') # Create a variable n3>=0
    n4 = pulp.LpVariable("n4", lowBound=0, upBound=None, cat='Continuous') # Create a variable n4>=0
    n5 = pulp.LpVariable("n5", lowBound=0, upBound=None, cat='Continuous') # Create a variable n5>=0
    n6 = pulp.LpVariable("n6", lowBound=0, upBound=None, cat='Continuous') # Create a variable n6>=0
    n7 = pulp.LpVariable("n7", lowBound=0, upBound=None, cat='Continuous') # Create a variable n7>=0
    n8 = pulp.LpVariable("n8", lowBound=0, upBound=None, cat='Continuous') # Create a variable n8>=0
    n9 = pulp.LpVariable("n9", lowBound=0, upBound=None, cat='Continuous') # Create a variable n9>=0
   
    model += m1*data["total score"][k]  #objective function
    
    model += n1*data["Total energy consumption(kWh)"][k] + n2*data["total water consumption (m3)"][k] + n3*data["total volume of waste water(m3)"][k]+ n4*data["Percentage modal split for commuting by staff (bus)"][k] + n5*data["Percentage modal split for commuting by students (bus)"][k]+ n6*data["Teaching students headcount"][k] + n7*data["Research students headcount"][k]+ n8*data["public lectures"][k] + n9*data["UK government regeneration funds"][k] == 1 #first constraint (equality)

    for i in range(7): # loop for 4 times for the next 4 constraints
        model += m1*data["total score"][i]- n1*data["Total energy consumption(kWh)"][i] - n2*data["total water consumption (m3)"][i] - n3*data["total volume of waste water(m3)"][i]- n4*data["Percentage modal split for commuting by staff (bus)"][i] - n5*data["Percentage modal split for commuting by students (bus)"][i]- n6*data["Teaching students headcount"][i] - n7*data["Research students headcount"][i]- n8*data["public lectures"][i] - n9*data["UK government regeneration funds"][i] <= 0
   
    status = model.solve(GLPK(msg=True, options=['--ranges', 'sensitivity.txt'])) # Solve the Problem and get the sensitivity analysis report
    print("Model Status:{}".format(pulp.LpStatus[model.status])) # Print the status of the solution

    # To Display Optimal Decision Variables & Reduced Cost per Variable
    for v in model.variables():
          print(v.name, "=", v.varValue, "\tReduced Cost =", v.dj)
  

    # To Display Optimal Value of Objective Function
    print("Objective=", pulp.value(model.objective))

    # To Display Shadow Price & Slack Value per Constraint
    print ("\nSensitivity Analysis\nConstraint\t\tShadow Price\tSlack")
    for name, c in model.constraints.items():
          print(name, ":", c, "\t", c.pi, "\t\t", c.slack)
```
